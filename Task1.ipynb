##### Feature Scaling

Scaled `Time` and `Amount` using `StandardScaler` to match the PCA-transformed features.


```python
from sklearn.preprocessing import StandardScaler
```

```python
df['Scaled_Amount'] = StandardScaler().fit_transform(df['Amount'].values.reshape(-1,1))
df['Scaled_Time'] = StandardScaler().fit_transform(df['Time'].values.reshape(-1,1))
```

```python
# drop original columns
df = df.drop(['Time','Amount'],axis=1)
df.head()
```

```python
# rearrangin columns
columns = [col for col in df.columns if col != 'Class'] + ['Class']
df = df[columns]
df.head()
```

```python
# final check
df.describe()
df['Class'].value_counts()
```

The dataset is highly imbalanced. Only 492 out of 284,807 transactions are fraudulent.


```python
df.to_csv('Cleaned_CreditCard.csv',index=False)
```
